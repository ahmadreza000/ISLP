```python
import numpy as np

np.random.seed(1)

# a
x = np.random.normal(0,1,100)
# b
eps = np.random.normal(0,0.25,100)
# c
Y = -1 + 0.5*x + eps
len = Y.shape[0]
print("length of Y is ",len)
# d
import matplotlib.pyplot as plt
plt.scatter(x, Y)
plt.show()
```

```python
# e
x_mean = x.mean()
Y_mean = Y.mean()
num = 0
denum = 0
for i in range(len):
    num = num + (x[i]-x_mean)*(Y[i]-Y_mean)
    denum = denum + ((x[i]-x_mean)**2)
beta1 = num/denum
beta0 = Y_mean - (beta1*x_mean)
print("beta0 = ",beta0 ," \nbeta1 = ",beta1)
```

```python
import statsmodels.api as sm

model = sm.OLS(Y, x)
results = model.fit()
print(results.summary())
```

```python
# f
plt.scatter(x, Y)
plt.axline((0,beta0), slope=beta1, color = 'red')
plt.axline((0,-1), slope=0.5, color = 'yellow')
plt.legend(['Population','Y_hat','Y'])
plt.show()
```

```python
# g
a = np.array([[len,x.sum(),(x**2).sum()],[x.sum(),(x**2).sum(),(x**3).sum()],[(x**2).sum(),(x**3).sum(),(x**4).sum()]])
b = np.array([[Y.sum()],[(Y*x).sum()],[(Y*(x**2)).sum()]])
ans = np.linalg.inv(a).dot(b)
ans
```
